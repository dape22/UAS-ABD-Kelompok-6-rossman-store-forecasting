# Preprocessing

```python
# Integrasi karakteristik detail setiap toko dari dataset metadata 
train = train.merge(
    metadata[["Store", "StoreType", "Assortment","Promo2","Promo2SinceWeek","Promo2SinceYear","PromoInterval","CompetitionDistance","CompetitionOpenSinceMonth","CompetitionOpenSinceYear"]],
    on="Store",
    how="left"
)

test = test.merge(
    metadata[["Store", "StoreType", "Assortment","Promo2","Promo2SinceWeek","Promo2SinceYear","PromoInterval","CompetitionDistance","CompetitionOpenSinceMonth","CompetitionOpenSinceYear"]],
    on="Store",
    how="left"
)

train.head()
```

```python
day_map = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday",
    6: "Saturday",
    7: "Sunday"
}

open_map = {
    0: "Closed",
    1: "Open"
}

state_map = {
    "0": "Nothing",
    0: "Nothing",
    "a": "Public Holiday",
    "b": "Easter Holiday",
    "c": "Christmas"
}

school_map = {
    0: "No School Holiday Effect",
    1: "School Holiday Effect",
    "0": "No School Holiday Effect",
    "1": "School Holiday Effect"
}

storetype_map = {
    "a": "Store Type A",
    "b": "Store Type B",
    "c": "Store Type C",
    "d": "Store Type D"
}

assort_map = {
    "a": "Basic",
    "b": "Extra",
    "c": "Extended"
}


daily_promo_map = {0: "No Daily Promo", 1: "Daily Promo Active"}
seasonal_promo_map = {0: "No Seasonal Promotions", 1: "Seasonal Promotions"}

## Konversi nama hari
train["DayOfWeek"] = train["DayOfWeek"].map(day_map)
test["DayOfWeek"] = test["DayOfWeek"].map(day_map)

## Konversi buka tutup toko
train["Open"] = train["Open"].map(open_map)
test["Open"] = test["Open"].map(open_map)

## Konversi jenis jenis libur
train["StateHoliday"] = train["StateHoliday"].map(state_map)
test["StateHoliday"] = test["StateHoliday"].map(state_map)

## Konversi hari libur sekolah
train["SchoolHoliday"] = train["SchoolHoliday"].map(school_map)
test["SchoolHoliday"] = test["SchoolHoliday"].map(school_map)

## Konversi kategori promo
train["Promo"] = train["Promo"].map(daily_promo_map)
test["Promo"] = test["Promo"].map(daily_promo_map)

train["Promo2"] = train["Promo2"].map(seasonal_promo_map)
test["Promo2"] = test["Promo2"].map(seasonal_promo_map)

## Konversi tipe toko
train["StoreType"] = train["StoreType"].map(storetype_map)
test["StoreType"]  = test["StoreType"].map(storetype_map)

## Konversi tipe level toko
train["Assortment"] = train["Assortment"].map(assort_map)
test["Assortment"]  = test["Assortment"].map(assort_map)

train.head()
```

```python
## ======= Mapping bulan Promo2 
month_map = {
    "Jan":1,"Feb":2,"Mar":3,"Apr":4,
    "May":5,"Jun":6,"Jul":7,"Aug":8,
    "Sept":9,"Oct":10,"Nov":11,"Dec":12
}

## ======= Fungsi mengubah PromoInterval → list bulan angka
def get_month_list(x):
    if pd.isna(x):
        return []
    return [month_map[m] for m in x.split(",")]

train["Promo2Months"] = train["PromoInterval"].apply(get_month_list)
test["Promo2Months"]  = test["PromoInterval"].apply(get_month_list)


## ======= Buat tanggal mulai Promo2 (Year + Week → datetime)
train["Promo2Start"] = pd.to_datetime(
    train["Promo2SinceYear"].astype("Int64").astype(str) + "-" +
    train["Promo2SinceWeek"].astype("Int64").astype(str) + "-1",
    format="%Y-%W-%w",
    errors="coerce"
)

test["Promo2Start"] = pd.to_datetime(
    test["Promo2SinceYear"].astype("Int64").astype(str) + "-" +
    test["Promo2SinceWeek"].astype("Int64").astype(str) + "-1",
    format="%Y-%W-%w",
    errors="coerce"
)


## ======= Pastikan kolom Date bertipe datetime
train["Date"] = pd.to_datetime(train["Date"], errors="coerce")
test["Date"]  = pd.to_datetime(test["Date"], errors="coerce")


## ======= Default: tidak ada promo2 aktif
train["PromoSeasonalActive"] = "Not Active"
test["PromoSeasonalActive"]  = "Not Active"


## ======= Tentukan baris yang berpotensi Promo2 aktif
mask_train = (train["Promo2"] == "Seasonal Promotions") & \
             (train["Date"] >= train["Promo2Start"])

mask_test = (test["Promo2"] == "Seasonal Promotions") & \
            (test["Date"] >= test["Promo2Start"])


## ======= Tandai Promo2 aktif jika bulan transaksi ada dalam Promo2Months
train.loc[mask_train, "PromoSeasonalActive"] = train.loc[mask_train].apply(
    lambda x: "Active" if x["Date"].month in x["Promo2Months"] else "Not Active",
    axis=1
)

test.loc[mask_test, "PromoSeasonalActive"] = test.loc[mask_test].apply(
    lambda x: "Active" if x["Date"].month in x["Promo2Months"] else "Not Active",
    axis=1
)

train.head()
```

```python
import numpy as np
## ======= Competition Distance Handling
# Mengisi missing value CompetitionDistance
# Jika jarak kompetitor tidak diketahui → anggap sangat jauh (nilai maksimum)
train["CompetitionDistance"].fillna(train["CompetitionDistance"].max(), inplace=True)
test["CompetitionDistance"].fillna(test["CompetitionDistance"].max(), inplace=True)


## ======= Buat tanggal mulai kompetitor aktif (AMAN)
def build_competition_date(df):
    df["CompetitionOpenSinceYear"]  = df["CompetitionOpenSinceYear"].astype("float")
    df["CompetitionOpenSinceMonth"] = df["CompetitionOpenSinceMonth"].astype("float")

    df["CompetitionStart"] = pd.to_datetime(
        dict(
            year=df["CompetitionOpenSinceYear"].fillna(0).astype(int),
            month=df["CompetitionOpenSinceMonth"].fillna(1).astype(int),
            day=1
        ),
        errors="coerce"   # kalau invalid → NaT
    )
    return df

train = build_competition_date(train)
test  = build_competition_date(test)


## ======= Pastikan kolom Date datetime
train["Date"] = pd.to_datetime(train["Date"], errors="coerce")
test["Date"]  = pd.to_datetime(test["Date"], errors="coerce")


## ======= Competition Is Active (Active / Not Active)
train["CompetitionIsActive"] = np.where(
    (train["CompetitionStart"].notna()) & (train["Date"] >= train["CompetitionStart"]),
    "Active",
    "Not Active"
)

test["CompetitionIsActive"] = np.where(
    (test["CompetitionStart"].notna()) & (test["Date"] >= test["CompetitionStart"]),
    "Active",
    "Not Active"
)

train.head()
```

# Penyimpanan data ke Elasticsearch

```python
from elasticsearch import Elasticsearch

es = Elasticsearch("http://localhost:9200")

print(es.info())
```

```python
records = data_train_final.to_dict(orient="records")
records2 = data_test_final.to_dict(orient="records")
```

```python
records
```

```python
from elasticsearch.helpers import bulk
actions_train = [
    {
        "_index": "uas-rossmann-train",
        "_source": doc
    }
    for doc in records
]

actions_test = [
    {
        "_index": "uas-rossmann-test",
        "_source": doc
    }
    for doc in records2
]

bulk(es, actions_train)
bulk(es, actions_test)
```
